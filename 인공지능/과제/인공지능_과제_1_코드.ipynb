```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Grid_Data.csv')
X = data['X'].values
Y = data['Y'].values

def loss_function(w, b, X, Y):
    predicted_Y = w * X + b
    return np.mean((Y - predicted_Y) ** 2)

def addictive_loss_function(w_range, b_range, iterations):
    best_w = 0
    best_b = 0
    best_loss = np.inf

    for epoch in range(iterations):
        w = np.random.uniform(w_range[0], w_range[1])
        b = np.random.uniform(b_range[0], b_range[1])

        current_loss = loss_function(w, b, X, Y)

        if current_loss < best_loss:
            best_loss = current_loss
            best_w = w
            best_b = b

        # if epoch%50 == 0:
        #     print(f"{epoch}번 ,w = {best_w}, b = {best_b}, 손실 = {best_loss}")

    return best_w, best_b, best_loss
```

```python
# w : 0 ~ 10, b : -5 ~ 5, 반복횟수 : 100
w_range1 = (0,10)
b_range1 = (-5,5)
iterations1 = 100
best_w1, best_b1, best_loss1 = addictive_loss_function(w_range1, b_range1, iterations1)

print(f"{iterations1}번, w = {best_w1}, b = {best_b1}, 손실 = {best_loss1}")
```

```python
total_loss = 0
total_w = 0
total_b = 0
for count in range(10):

    # w : 0 ~ 10, b : -5 ~ 5, 반복횟수 : 100
    w_range1 = (0,10)
    b_range1 = (-5,5)
    iterations1 = 100
    best_w1, best_b1, best_loss1 = addictive_loss_function(w_range1, b_range1, iterations1)
    total_loss += best_loss1
    total_w += best_w1
    total_b += best_b1

    print(f"{count+1}번, w = {best_w1}, b = {best_b1}, MSE = {best_loss1}")
print(f"평균 w = {total_w/10}, 평균 b = {total_b/10}, 평균 MSE = {total_loss/10}")
```

```python
plt.scatter(X, Y, color='blue')
predicted_Y = (total_w/10) * X + total_b/10
plt.grid()
plt.plot(X, predicted_Y, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(1.5, 11, "MSE = {:.3f}".format(total_loss/10))
plt.text(1.5, 10.5, "w = {:.3f}".format(total_w/10))
plt.text(1.5, 10, "b = {:.3f}".format(total_b/10))
plt.show()
```
